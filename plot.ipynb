```python
import matplotlib.pyplot as plt
```

```python
terms =range(1,7)
GPA=[3.38,3.15,3.2,3.74,3.82,4]
```

```python
plt.plot(terms,GPA,color='red',marker='o',linestyle='dotted')
plt.xlabel('Term')
plt.ylabel('GPA')
plt.title('GPA per term')
fig = plt.figure()
fig.savefig(fname='first plot')

```
