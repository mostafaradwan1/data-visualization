```python
import matplotlib.pyplot as plt
```

```python
books=['data science from scratch','hands on machine learning','data visualization','deep learning from scratch']
```

```python
num_pages=[500,800,200,450]
```

```python
plt.bar(books,num_pages)
plt.xlabel('books')
plt.ylabel('number of pages')
```

```python
actors = ["omar elsherif", "adel emam", "hasan hosny", "ahmed zaki"]
num_prizes = [9, 11, 12,  10]
```

```python
plt.bar(range(len(actors)),num_prizes,2,edgecolor=(0,0,0))
plt.xticks(range(len(actors)),actors)
plt.xlabel('actors')
plt.ylabel('num prizes')
plt.title('prizes for actors')
```
